# Average Prices of Vegetables and Fruits in Delhi

```python
import pandas as pd

import matplotlib.pyplot as plt



vegetables_data = {

    2005: 0.30,

    2010: 0.40,

    2015: 0.50,

    2020: 0.60

}



fruits_data = {

    2005: 0.80,

    2010: 0.50,

    2015: 0.60,

    2020: 1.20

}



plt.figure(figsize=(10, 6))

plt.plot(list(vegetables_data.keys()), list(vegetables_data.values()), color='green', label='Vegetables', marker='o')

plt.plot(list(fruits_data.keys()), list(fruits_data.values()), color='orange', label='Fruits', marker='o')



plt.xticks([2005, 2010, 2015, 2020])  

plt.xlabel('Year')

plt.ylabel('Average Price (in USD)')

plt.title('Average Prices of Vegetables and Fruits in Delhi')

plt.legend()

plt.tight_layout()

plt.show()

```

# Delhi Vegetable and Fruit Prices Over the Years

```python
import pandas as pd
import matplotlib.pyplot as plt


years = [2015, 2016, 2017, 2018, 2019, 2020]
vegetable_prices = [20, 25, 30, 28, 32, 35]  
fruit_prices = [30, 35, 40, 38, 42, 45]  


plt.figure(figsize=(10, 6))
plt.plot(years, vegetable_prices, label='Vegetable Prices', marker='o')
plt.plot(years, fruit_prices, label='Fruit Prices', marker='o')
plt.xlabel('Year')
plt.ylabel('Price (in INR)')
plt.title('Delhi Vegetable and Fruit Prices Over the Years')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
